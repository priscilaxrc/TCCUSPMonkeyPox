```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/TCCPriscila/Teste_Daily_Country_Wise_Confirmed_Cases.csv',sep=';', index_col='Country', header='infer',parse_dates=True)
df=df.dropna()
print('Shape of data',df.shape)
df.head()
df
```

```python
dft = df.transpose()
dft
```

```python
print(dft['Spain'])

DataTest = dft['Spain']
```

```python
from statsmodels.tsa.arima.model import ARIMA
model = ARIMA(DataTest, order=(28, 2, 5))
fit_model = model.fit()
```

```python
arima_forecast = fit_model.forecast(steps=30)
print(arima_forecast)
```

```python
import matplotlib.pyplot as plt
# Plotar as curvas de previsão e real
plt.figure(figsize=(10, 6))
plt.plot(DataTest, label='Real')
plt.plot(arima_forecast, label='Previsão', color='red')
plt.xlabel('Período')
plt.ylabel('Valor')
plt.title('Previsão usando modelo ARIMA Spain')
plt.legend()
plt.show()
```
