```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as plt
```

```python
import pandas as pd
df=pd.read_csv('../content/drive/MyDrive/TCCPriscila/Teste_Daily_Country_Wise_Confirmed_Cases.csv',sep=';', index_col='Country', header='infer',parse_dates=True)
df=df.dropna()
print('Shape of data',df.shape)
df.head()
df
```

```python
dftn = df.transpose()
dft = dftn[:100]
dft
```

```python
print(dft['Spain'])
DataTest = dft['Spain']
```

```python
from statsmodels.tsa.arima.model import ARIMA
model = ARIMA(DataTest, order=(10, 0, 10))
fit_model = model.fit()
```

```python
arima_forecast = fit_model.forecast(steps=100)
print(arima_forecast)
```

```python
import matplotlib.pyplot as plt
# Plotar as curvas de previsão e real
plt.figure(figsize=(10, 6))
plt.plot(DataTest.values, label='Real')
plt.plot(arima_forecast.values, label='Previsão', color='red')
plt.xlabel('Período')
plt.ylabel('Valor')
plt.title('Previsão usando modelo ARIMA - Spain')
plt.legend()
plt.show()
```

```python
#Calculo de Erro
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error, median_absolute_error, max_error

MSE = mean_squared_error(DataTest.values,arima_forecast.values)
MAE = mean_absolute_error(DataTest.values,arima_forecast.values)
MAPE = mean_absolute_percentage_error(DataTest.values,arima_forecast.values)
MedAE = median_absolute_error(DataTest.values,arima_forecast.values)
RMSE = np.sqrt(MSE)
ME = max_error(DataTest.values,arima_forecast.values)

print("MSE = {:0.2f}".format(MSE))
print("MAE = {:0.2f}".format(MAE))
print("MAPE = {:0.2f}".format(MAPE))
print("MedAE = {:0.2f}".format(MedAE))
print("RMSE = {:0.2f}".format(RMSE))
print("ME = {:0.2f}".format(ME))
```
